## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python
def mark_outliers_as_nan(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), np.nan, df[column])

for column in df.columns:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':
        mark_outliers_as_nan(df, column)
```

```python
df.fillna(df['Light_Conditions'].mode(), inplace=True)
```

```python
df.fillna(df['Urban_or_Rural_Area'].mode(), inplace=True)
```

```python
df.fillna(df['Number_of_Casualties'].mean(), inplace=True)
```

```python
df.fillna(df['Accident_Severity'].mean(), inplace=True)
```

```python
df_scaler = df.select_dtypes(include=[np.number])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_scaler)

df_scaled = pd.DataFrame(df_scaled, columns=df_scaler.columns)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

imputer = SimpleImputer(strategy='mean')
df_imputed = imputer.fit_transform(df_scaled)

df_imputed = pd.DataFrame(df_imputed, columns=df_scaled.columns)

pca = PCA(n_components=0.9)
df_pca = pca.fit_transform(df_imputed)

df_pca = pd.DataFrame(df_pca)
```

## Clustering
Apply K-Means clustering. Determine the optimal number of clusters and justify your choice.
* Find the `n_clusters` parameter using the elbow method.
* Train the model.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_pca)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()

optimal_k = 3

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(df_pca)
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
import matplotlib.pyplot as plt

labels = kmeans.labels_
df['Cluster'] = labels

for i in range(optimal_k):
    cluster_data = df[df['Cluster'] == i]
    print(f"Cluster {i}:")
    print (cluster_data.describe())

```

## Plot the data points with their predicted cluster center

```python
import matplotlib.pyplot as plt

labels = kmeans.labels_

df_pca['Cluster'] = labels

from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(df_pca, labels)
print("The average silhouette_score is :", silhouette_avg)

pca_2d = PCA(n_components=2)
df_pca_2d = pca_2d.fit_transform(df_imputed)
plt.scatter(df_pca_2d[:, 0], df_pca_2d[:, 1], c=labels)
plt.title('Clusters Visualization')
plt.show()
```
